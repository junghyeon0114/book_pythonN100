**22)데이터프레임의 행 다루기(선택,추가,변경,삭제)**

- iloc : 숫자 기반

- loc : 명칭 기반

```python
import pandas as pd

# dictionary를 데이터프레임으로 변환
dict_data = {'c0':[1,2,3], 'c1':[4, 5, 6]}
df = pd.DataFrame(dict_data, index=['r0', 'r1', 'r2'])
print(df, "\n")

# 행 선택
row0 = df.iloc[0]
row1 = df.iloc[1]
row2 = df.loc['r2']

print(row0, "\n")
print(row1, "\n")
print(row2, "\n")

# 행 추가
df.loc['r3'] = 10, 20
print(df, "\n")

df.loc['r4'] = 0
print(df, "\n")

# 행 변경
df.loc['r3'] = df.loc['r4']
print(df, "\n")

# 행 삭제
df.drop('r4', axis=0, inplace=True)
print(df, "\n")

df.drop(['r1','r3'], axis=0, inplace=True)
print(df)
```

  2. interpolation='bilinear'
    **interpolation**은 이미지를 표시할 때 사용하는 보간법을 지정하는 매개변수
  - 'bilinear'는 양선형 보간법(Bilinear interpolation)을 의미하며, 두 점 사이에서 보간하여 중간 값을 계산합니다. 이 방법은 이미지가 부드럽고 흐릿한 느낌을 줄 수 있습니다.
  다른 옵션으로는 nearest, bicubic, spline16, spline36 등이 있으며, 각 보간법은 이미지가 어떻게 보일지에 영향을 줍니다. bilinear는 대부분의 경우 자연스럽고 부드러운 효과를 줍니다.


3. cmap='YlOrBr'
**cmap**은 색상 맵을 지정하는 매개변수로, 이미지를 색상으로 시각화할 때 색상 팔레트를 설정합니다.
'YlOrBr'는 Yellow-Orange-Brown 색상 팔레트를 의미하며, 색상 그라데이션이 노란색에서 주황색, 갈색으로 이어지는 색상 팔레트입니다.
matplotlib의 cmap 매개변수는 다양한 색상 팔레트를 지원하며, 이를 통해 워드클라우드의 시각적인 스타일을 조정할 수 있습니다.

```python
# 한글 글꼴 설치 (NanumGothic)
!apt-get install -y fonts-nanum

# 설치한 글꼴을 WordCloud에 적용
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from konlpy.tag import Hannanum
from collections import Counter

# 텍스트 파일 열기
text = open('/content/drive/MyDrive/5674-849/data/2018_president_message.txt', encoding='cp949').read()


# 한글 형태소 분석하기
engin = Hannanum()

# 길이가 1보다 큰 명사만 추출
nouns = engin.nouns(text)
nouns = [n for n in nouns if len(n) > 1]

# 단어 숫자 세기
count = Counter(nouns)

# 빈도수가 높은 상위 50개 항목을 리스트 형태로 반환 :단어-빈도 튜플형식
tags = count.most_common(50)

# 폰트 경로 확인
font_path = '/usr/share/fonts/truetype/nanum/NanumGothic.ttf'

# 워드 클라우드 이미지 생성하기
wordcloud = WordCloud(font_path=font_path,
                      background_color='white',
                      width=1200,
                      height=800).generate_from_frequencies(dict(tags))

# 화면에 출력하기
fig = plt.figure()
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

# 이미지 출력
plt.show()
```

**33) 워드클라우드(3) - 이미지 마스킹 처리하기**

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

# 텍스트 파일 열기
text = open('/content/drive/MyDrive/5674-849/data/usa_president_message.txt', encoding='UTF-8').read()

# 이미지 마스킹 처리
image_path = '/content/drive/MyDrive/5674-849/data/usa_map.jpg'
usa_map = np.array(Image.open(image_path))

# 워드 클라우드 이미지 생성하기
# stopwords=STOPWORDS :워드클라우드 생성 시 불용어(stopwords)를 제외하고, 나머지 중요한 단어만 포함시킨다는 의미
wordcloud = WordCloud(background_color='white',
                      max_font_size = 100,
                      max_words=1000,
                      stopwords=STOPWORDS,
                      mask=usa_map).generate(text)

# 화면에 출력하기
fig = plt.figure(figsize=(15,15))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')

# SVG 객체로 이미지 저장하기
#plt.savefig('./output/usa_president_message_wordcloud_with_map.svg')
```

**34) 레이더 차트 - 별 그림**

```python
import matplotlib.pyplot as plt
import pandas as pd
from math import pi

# 데이터  로딩
kbo = pd.read_csv("/content/drive/MyDrive/5674-849/data/kbo.csv")
print(kbo.head())
print("\n")

# 변수 지정
var = kbo.columns.to_list()[1:]
print(var)
print("\n")

# 1위 팀 값 지정
val1 = kbo.loc[0, :].drop('팀명').values.tolist()
val1 += val1[:1]
print(val1)
print("\n")

# 10위 팀 값 지정
val2 = kbo.loc[1, :].drop('팀명').values.tolist()
val2 += val2[:1]
print(val2)
print("\n")

# 변수별 각도 계산
num_var = len(var)

# 2pi/변수개수를 라디안으로 표현
deg = [n / float(num_var) * 2 * pi for n in range(num_var)]
deg += deg[:1]
print(deg)
print("\n")

# matplotlib 한글 폰트 오류 문제 해결
from matplotlib import font_manager, rc
font_path = "/content/drive/MyDrive/5674-849/data/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 레이더 차트 그리기
ax = plt.subplot(111, polar=True)
plt.xticks(deg[:-1], var, color='grey', size=10)

# 45도 각도로 y값 빨간색으로 표시 : 25,50,75,100
ax.set_rlabel_position(45)
plt.yticks([25,50,75,100], ["25","50","75","100"], color="red", size=7)
plt.ylim(0,100)

#1위팀
ax.plot(deg, val1, linewidth=1, linestyle='solid', label='1위 팀')
ax.fill(deg, val1, 'orange', alpha=0.2)

# 2위침
ax.plot(deg, val2, linewidth=1, linestyle='solid', label='10위 팀')
ax.fill(deg, val2, 'blue', alpha=0.2)

plt.legend(loc='best', bbox_to_anchor=(0.05, 0.95))
```

**35) 히트맵 그리기**

- sns.heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Seaborn 제공 데이터셋 가져오기
df = sns.load_dataset('iris')
print(df.head())
print("\n")
print(df.columns.values)

# 상관계수를 구해서 히트맵으로 그리기
plt.figure(figsize=(10,10))
corr = df.loc[:,'sepal_length':'petal_width'].corr()
print(corr.head())
print("\n")

# 히트맵 그리기
sns.set(font_scale=1.5)
sns.heatmap(corr,
            annot=True,
            cmap='PuBuGn',
            fmt='.1f',
            square=True,
            linewidth=0.5,
            cbar=False)

plt.show()
```

```python
import pandas as pd
import calmap
import matplotlib.pyplot as plt

# 표현 가능한 최대 열 개수
pd.set_option('display.max_columns', 20)

# KOSPI 데이터 준비
#  parse_dates=['년/월/일'] : 열을 날짜 형식으로 인식
df = pd.read_excel('/content/drive/MyDrive/5674-849/data/kospi.xls', parse_dates=['년/월/일'])
print(df.head())
print("\n")

df.columns = ['date', 'price', 'up_down', 'change', 'start', 'high', 'low',
              'vol_num', 'vol_amt', 'mkt_cap']
df = df.set_index('date', drop=True)
print(df.head())
print("\n")

# Calendar Map 표현
# monthticks=1 : 1~12 월 간격. 1로 설정하면 모든 월을
# daylabels : 월~일 요일 표현 방식
# dayticks : 요일 표현 간격 (0,2,4,6 = 월,수,금,일)
#  cmap : 컬러맵
# linewidth : 구분 선의 두깨
# fillcolor : 누락값
# fig_kws : 채색 달력 그래프 크기
# subplot_kws : 연도 표시 형식
plt.figure(figsize=(16,8))
calmap.calendarplot(df.change,
                    monthticks=1, daylabels='MTWTFSS', dayticks=[0, 2, 4, 6],
                    cmap='YlGn', linewidth=0.05, fillcolor='grey',
                    fig_kws=dict(figsize=(14, 6)),
                    yearlabel_kws=dict(color='black', fontsize=12),
                    subplot_kws=dict(title='2018 KOSPI Price Trend'),
                    )

plt.show()
```
